1. Escriba un código que sume los números múltiplos de 17 que hay entre 0 y 100,000, es decir, que sume 0 + 17 + 34 + 51 + ...

````python
from numpy import *


# Inicio la variable suma para almacenar la suma de los múltiplos de 17
suma = 0

# Itero sobre los numeros del 0 al 100,000     
for i in range(0, 100001):
    # Verifico si el número actual em i es un múltiplo de 17
    if i % 17 == 0:
        # Si es un múltiplo de 17, lo sumamos a la variable suma
        suma += i

# Imprimo la suma de los multiplos de 17
print("La suma de los múltiplos de 17 entre 0 y 100,000 es: ", suma)


````

2. Haga la gráfica de la función $f(x)=1 + \frac{x^2}{2!} + \frac{x^4}{4!}$ y de la función $\cosh (x)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo de $-1$ a $1$. Explique las similitudes y diferencias.

````python
import numpy as np
from matplotlib.pyplot import *
import math as m
# importo las librerias necesarias 


#requiero que ambas funciones se encuentren en la misma gráfica y en el mismo intervalo


#defino la función 
def f(x):
    return 1 + (x**2 / m.factorial(2)) + (x**4 / m.factorial(4))

x = np.linspace(-1, 1, 1000) # Definp el intervalo [-1, 1]
y = f(x) # calculo la funcion f(x)


cosh_y = np.cosh(x)  # calculo la funcion cosh(x)

'''
p1, p2 =plot(x, y, x, cosh_y)
xlabel("x")
ylabel("y")
title("Comparación de f(x) y cosh(x)") '''
````

````python
#grafico las funciones


plt.plot(x, y, label='f(x) = 1 + x^2/2! + x^4/4!')
plt.plot(x, cosh_y, label='cosh(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Comparación de f(x) y cosh(x)")
plt.legend()

plt.show()
    
````

3. Realice una función que simule un dado de 8 caras. Es decir, haga una función que proporcione un número entre 1 y 8. Por ejemplo:
```python
dado()
```
y a la salida, podría proporcionar el número: ```5```.

````python
import  numpy as np 
#importo las libreria necesaria


def mi_dado():
    dado= str(np.random.rand()) #tomo un numero aleatorio co la funcion rand() 
    contador=0
    for n in range(2, len(dado)):
        cara = int(dado[n])
        if cara <= 7: #limito a que el resultado sea menor igual a 8
            print(cara+1)   
            return 
        else: contador += 1 
        return

mi_dado()


````

4. Realice una función que reciba un número telefónico de 10 dígitos y que a la salida proporcione la suma y el producto de todos de todos sus dígitos distintos de cero. Compruebe el funcionamiento de su función con algún teléfono. Ejemplo:
```
In: sumar_tel(5545973256)
Out: Suma = 51 ; Producto = 5,670,000
```

````python

def sumar_tel(numero):
    # Convertimos el número a cadena para poder manejarlo mejor 
    numero_str = str(numero)
    

    # Inicio las variables suma y multiplicación para almacenar los resultados
    suma = 0
    multiplicacion = 1

    # Itero sobre cada digito del numero
    for digito in numero_str:
        # convierto el digito a entero
        digito_int = int(digito)
        # Sumamos el dígito a la suma acumulada
        suma += digito_int
        
        
        # Si existe un cero dentro del numero telefonico lo omitimos:   
        if digito_int==0:
            multiplicacion *= 1  
        else:
            multiplicacion *= digito_int


    return suma, multiplicacion

````

````python

# Ejemplo 
numero = int(input("Ingrese numero de telefono: ")) 
resultado_suma, resultado_multiplicacion = sumar_tel(numero)
print("Suma=", resultado_suma)
print("producto=", resultado_multiplicacion)
````
