1. Realice una función que reciba un número y que regrese ese mismo número con cambio de signo pero
con sus cifras reflejadas y conservando la posición del número decimal. Por ejemplo, si el número
proporcionado es 8589.42, la función deberá regresar -2498.58.


```python
import numpy as np

def invertir_numero(numero):
    # Convierto el número a cadena para facilitar su manipulación
    numero_str = str(numero)

    # Determinar si el número es negativo
    es_negativo = False
    if numero_str[0] == '-':
        es_negativo = True
        numero_str = numero_str[1:]  # Eliminar el signo negativo

    # Encontrar la posición del punto decimal si existe
    punto_decimal_pos = len(numero_str) if '.' not in numero_str else numero_str.index('.')

    # Eliminar el punto decimal y luego invertir la cadena
    numero_sin_punto = numero_str.replace('.', '')[::-1]

    # Inserto el punto decimal en la posición correspondiente después de invertir la cadena
    numero_invertido = numero_sin_punto[:punto_decimal_pos] + '.' + numero_sin_punto[punto_decimal_pos:]

    # Agrego el signo negativo si es necesario
    if es_negativo:
        numero_invertido = numero_invertido
    else:
        numero_invertido = '-' + numero_invertido

    return numero_invertido

# Ejemplos
numero_flotante = 8589.42
numero_entero = -562

print("Número original:", numero_flotante)
print("Salida:", invertir_numero(numero_flotante))

print("Número original:", numero_entero)
print("Salida:", invertir_numero(numero_entero))
```

      1 & \text{si } x \in [n,n+1) \text{ con } n \text{ par} \\
      0 & \text{si } x \in [n,n+1) \text{ con } n \text{ impar}
   \end{cases}$$

   donde n es la parte entera de x.

```python

import numpy as np
import matplotlib.pyplot as plt # importo las librerias necesarias


def mi_f(x): # Defino la función f(x)
  
    n = np.floor(x)   # Obtengo la parte entera de x
 # Necesitamos obtener la parte entera de x para determinar en qué intervalo [n, n+1) se encuentra x, 
    #esto nos permitirá asignar el valor de 1 a f(x) si n es par, y el valor de 0 si n es impar
    
    #ya que obtengo la parte entera de x,  ahora necesito verificar si n es par o impar
    if n % 2 == 0:
        return 1  # Si n es par, regresa 1
    else:
        return 0  # Si n es impar, regresa 0

# Generar un conjunto de puntos en el intervalo [-10, 10] para graficarla
x_valor = np.linspace(-10, 10, 1000)
y_valor = []
for x in x_valor:
        y_valor.append(mi_f(x))  # Calculo los valores de la función para cada punto y los agrego en la lista y_valor

# Grafico la funcion
plt.plot(x_valor, y_valor, label='$f(x)$')  # Grafica la función
plt.title('Gráfica de la función $f(x)$')
plt.xlabel('x')  
plt.ylabel('$f(x)$')  
plt.ylim(-0.1, 1.1)  # Limitar el rango del eje y entre -0.1 y 1. para visualizarla mejor
plt.legend() 
plt.show()  
```
