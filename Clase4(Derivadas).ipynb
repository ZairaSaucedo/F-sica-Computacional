```python
import numpy as np
from pylab import *
```

```python
def deriv_h5(x,y):
    h12 = 12*(x[1]-x[0])
    dv= zeros_like(y)
    for i in range(2, len(y)-2):
        dv[i] = -y[i+2] +8*y[i+1] -8*y[i-1] +y[i-2] #puntos inter 
    dv[0]=-25*y[0] +48*y[1] -36*y[2] +16*y[3] -3*y[4] # dy en i=0
    dv[1]=-25*y[1] +48*y[2] -36*y[3] +16*y[4] -3*y[5] # dy en i=1
    dv[-1]=25*y[-1] -48*y[-2] +36*y[-3] +16*y[-4] + 3*y[-5] # dy en i=0
    dv[-2]=25*y[-2] -48*y[-3] +36*y[-4] +16*y[-5] + 3*y[-6]
   
    dv= dv/h12
    return dv
    
```

```python
def deriv_h3(x,y):
    h2 = 2*(x[1]-x[0])
    dv= zeros_like(y)
    for i in range(2, len(y)-2):
        dv[i] = -y[i+1]-y[i-1] #puntos inter 
    dv[0]=-3*y[0]+4*y[1]-y[2] # dy en i=0
    dv[0]=-3*y[-1]-4*y[-2]+y[-3] 
    dv= dv/h2
    return dv
    
```

```python

x = linspace(-pi,pi, 200)
y = sin(x**2)
dv=2*x*cos(2**x)
dv_h5=deriv_h5(x,y)
dv_h3=deriv_h3(x,y)
plot(x,dv,label='derivada')
plot(x,dv_h5,label ='derivada $h^5$')
plot(x,dv_h3,label ='derivada $h^3$')

legend()
show()
```

```python
error_h5 = abs(dv-dv_h5)
error_h3= abs(dv-dv_h3)
print('Max err h3 =', error_h3[2:-2].max())
print('Max err h5 =', max(error_h5[2:-2]))
plot(x, error_h3, label = 'err $h^3$')
plot(x, error_h5, label = 'err $h^5$')
legend()
show()
```
