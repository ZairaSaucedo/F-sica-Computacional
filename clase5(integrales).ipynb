La expresión para la integral en el punto $x_i$ con el intervalo simétrico $\pm h$ es: 

$$ \int _{x_i-h} ^{x_i-h} f(x) dx = \left( \frac{f_{i-1}+4f{i} +f_{i+1} } {3}  \right) h $$

En caso de un intervalo par, hay que sumar:

$$ \int _{x_0} ^{x_1} f(x) dx = \left( \frac{23 f_0+26 f_1 -4f_2+6f_3-3f_4 } {48}  \right) h $$
